```python
#Develop
def split_file(file_content, chunk_size):
  chunks = [file_content[i:i + chunk_size]
    for i in range(0, len(file_content), chunk_size)]
  return chunks

#usage
file_content="aisffaeidfjgsdfgsdfsjdhfsdjhfafasfasfags fasjhags fjasfhgasf asfhgLERYWAEGFWAE FWAEFAWGEFIAGWEUOFYWEFUWYEFWGE FWEFYWGUFUGWYFASYVSUDHYBASUDFUSDYFASOHDFAJFHSDJFHSDJFSDHGFJHGSJDG SDFJSHDGF SDJFHGSD FJSDHF SJSHGD FJS FASDJHFGS DFJAOGFSDA FASHDFGS DFJAS ASDFJHADSG FASJ FASDFJHGAS DFSJ AHSDFJHDGS FASDHFGAS DFASJDHF SDFKJGHASDJ FSAJ FHADFJ AFSD"
chunk_size = 7
chunks = split_file(file_content, chunk_size)

for i, chunk in enumerate(chunks):
  print(f"Node{i + 1}: {chunk}")
```
